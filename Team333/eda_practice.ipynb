```python
place_df['범죄중분류'] = place_df['범죄중분류'].replace("기타 강간_강제추행등", "기타 강간 강제추행등")
place_df['범죄중분류'] = place_df['범죄중분류'].replace("절도범죄", "절도")
place_df['범죄중분류'] = place_df['범죄중분류'].replace("체포_감금", "체포 감금")
place_df['범죄중분류'] = place_df['범죄중분류'].replace("약취?유인", "약취 유인")
place_df['범죄중분류'] = place_df['범죄중분류'].replace("문서_인장", "문서 인장")
```
